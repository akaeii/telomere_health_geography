```python
import pingouin as pg
import pandas as pd
```

Loading Data

```python
age_data = pd.read_csv("age.csv")
```

```python
age_data
```

Converting Data into Long Format (One Row One Observation)

```python
age_data_cleaned = age_data.melt(var_name='Age Group', value_name='Telomere Length').dropna()
```

```python
age_data_cleaned
```

Descriptive Statistics

```python
age_data.describe()
```

```python
age_data_cleaned.describe()
```

Kruskal-Wallis (H-Test)

```python
kw = pg.kruskal(data=age_data_cleaned, dv="Telomere Length", between="Age Group", detailed=True)
```

```python
kw
```

H-Statistic → η² → Cohen's f

```python
kw_H = kw['H'].squeeze()
kw_n2 = (((kw_H - 5) + 1)/(615-5))
kw_cohens_f = ((kw_n2/(1-kw_n2))**0.5)
```

```python
print(f"Cohen's f of the test is: {kw_cohens_f.round(3)}")
print(f"Computed Power is: 0.15 (sample size adjusted to 616) (G*Power)")
```
